##### Following the results of Experiment 3, where increasing the total_skips threshold significantly improved packing efficiency but also caused the mean diameter of placed circles to drift below the intended mean of 40, we now turn to a key question: Is this downward shift in mean diameter an inherent limitation of the algorithm, or a byproduct of the constrained sheet size? To investigate this, Experiment 4 increases the sheet size from 500×500 to 1000×1000 while keeping all other parameters constant, including the diameter distribution and range. The goal is to see whether a larger sheet helps the algorithm place circles more fairly especially larger ones so that the original diameter distribution is better preserved, without shifting the mean or distorting its shape.

```python
#Experiment No.4

#random allocation of circle on a shaeet without overlapping
#distribution of circles (Normal Distribution)
## mean diameter = 40
## std of diameter = 10
## lower limit on diameter = 0
## higher limit on diameter = 80

# dimension of the sheet
# 1000 X 1000
```

```python
from utility_functions import generate_diameters, place_circles_bc1, save_circles, save_attempts, calculate_packing_efficiency, place_circles_bc1_kdtree, is_overlapping
from graphing_tools import plot_circles, plot_distribution
#Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import truncnorm
from collections import defaultdict
import time
```

```python
# --- Simulation Parameters ---


# Sheet dimensions
sheet_width = 1000
sheet_height = 1000

# Circle Distribution parameters
mean_diameter = 40
std_diameter = 10
min_diameter = 0
max_diameter = 80

population_size = 1000000          # Number of diameters to generate
max_attempts = 1000000              # Max attempts per circle before skipping

allocated_circles = []             # List of successfully placed circles
attempts_dict = defaultdict(int)   # Tracks how many attempts it took to place each circle
```

```python
total_skips = 100
print(time.time())
raw_diameters = generate_diameters(population_size, mean_diameter, std_diameter)
truncated_diameters = raw_diameters[(raw_diameters >= min_diameter) & (raw_diameters <= max_diameter)]
allocated_circles, attempts_dict = place_circles_bc1(truncated_diameters, total_skips, max_attempts, sheet_width, sheet_height)
print(time.time())
```

```python
save_circles(allocated_circles, 'data/allocated_circles_Experiment4a.csv')
save_attempts(attempts_dict, 'data/attempts_distribution_Experiment4a.csv')
```

##### We ran five sub-experiments (4a to 4e) with increasing total_skips values, and observed that packing efficiency continued to improve from 59.60% to 69.66%, echoing the trend from earlier experiments. However, the histograms of allocated diameters showed that the bias toward smaller circles persisted, even in the larger system. This indicates that while increased canvas size provides more flexibility and overall space, it does not fully eliminate the tendency of the algorithm to favor smaller circles during late-stage placements. Larger circles still struggle to find feasible locations as the simulation progresses, and without an explicit mechanism to preserve the diameter distribution, the mean naturally drifts downward.
