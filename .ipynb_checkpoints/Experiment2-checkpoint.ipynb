##### In this experiment, we aim to improve the computational efficiency of simulating non-overlapping random circle packing on a 2D canvas. In Experiment 1, a large sheet size (5000 x 5000) made the simulation memory-intensive and slow due to the large number of overlap checks and the high number of circle placement attempts. Therefore, in this follow-up experiment, we reduce the canvas size and run three sub-experiments with different standard deviations in the circle diameter distribution to observe how variance affects the circle packing efficiency.

```python
#Experiment No.2

#random allocation of circle on a shaeet without overlapping
#distribution of circles (Normal Distribution)
## mean diameter = 40
## std of diameter = 10
## lower limit on diameter = 0
## higher limit on diameter = 80

# dimension of the sheet
# 500 X 500
```

```python
#Import Libraries
from utility_functions import calculate_packing_efficiency
from graphing_tools import plot_circles, plot_distribution
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import truncnorm
from collections import defaultdict
```

```python
#Simulate a wide range of circle sizes (many small, some large)

# --- Simulation Parameters ---


# Sheet dimensions
sheet_width = 500
sheet_height = 500

# Circle Distribution parameters
mean_diameter = 40
std_diameter = 10
min_diameter = 0
max_diameter = 80


population_size = 1000000          # Number of diameters to generate
total_skips = 100                  # Number of failed full tries before exiting
max_attempts = 250000              # Max attempts per circle before skipping

allocated_circles = []             # List of successfully placed circles
attempts_dict = defaultdict(int)   # Tracks how many attempts it took to place each circle

```

```python
# Function to sample from a truncated normal distribution
# Step 1: Generate initial diameters from normal distribution (no truncation yet)
def generate_diameters(n):
    np.random.seed(42)
    return np.random.normal(loc=mean_diameter, scale=std_diameter, size=n).astype(int)

# Check for overlap
def is_overlapping(new_circle, existing_circles):
    x1, y1, r1 = new_circle
    for x2, y2, r2 in existing_circles:
        dist_sq = (x1 - x2)**2 + (y1 - y2)**2
        min_dist = (r1 + r2)**2
        if dist_sq < min_dist:
            return True
    return False

def place_circles(truncated_diameters):
  np.random.seed(42)
  total_tries = 0
  while total_tries < total_skips:
    diameter = np.random.choice(truncated_diameters)
    # print(diameter)
    attempts = 0

    while attempts < max_attempts:
      radius = diameter / 2
      x = int(np.random.uniform(radius, sheet_width - radius))
      y = int(np.random.uniform(radius, sheet_height - radius))
      new_circle = (x, y, radius)
      if not is_overlapping(new_circle, allocated_circles):
          allocated_circles.append(new_circle)
          attempts_dict[attempts] += 1
          break

      attempts += 1
    if attempts == max_attempts:
      total_tries += 1
      print(total_tries)

    # --- Save allocated_circles to a CSV file ---
    circle_df = pd.DataFrame(allocated_circles, columns=['x', 'y', 'radius'])
    circle_df.to_csv('data/allocated_circles_Experiment2a.csv', index=False)
    print("✅ Saved allocated circles to 'allocated_circles.csv'")

    # --- Save attempts_dict to a CSV file ---
    attempts_df = pd.DataFrame(list(attempts_dict.items()), columns=['attempts', 'count'])
    attempts_df.sort_values(by='attempts', inplace=True)
    attempts_df.to_csv('data/attempts_distribution_Experiment2a.csv', index=False)
    print("✅ Saved attempts distribution to 'attempts_distribution.csv'")

```

```python
raw_diameters = generate_diameters(population_size)
truncated_diameters = raw_diameters[(raw_diameters >= min_diameter) & (raw_diameters <= max_diameter)]

place_circles(truncated_diameters)
```

##### In Experiment 2, we successfully ran simulations for random, non-overlapping circle placement on a significantly smaller canvas (500x500) compared to our initial attempt in Experiment 1. By varying the standard deviation (σ) of the circle diameter distribution while keeping the mean diameter (μ = 40) and diameter range [0, 80] fixed, we were able to observe clear trends in packing efficiency. The results indicate that higher variability in circle size leads to better packing efficiency, with Experiment 2a (σ = 10) achieving the highest packing efficiency of 61.69%, followed by 2b (σ = 5) at 58.19%, and 2c (σ = 1), the most uniform case at only 54.64%.

##### This trend highlights an important geometric insight: a diverse range of circle sizes helps fill gaps more effectively.
