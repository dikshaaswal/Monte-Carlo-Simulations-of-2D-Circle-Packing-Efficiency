```python
#Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Generates 10 million random integers uniformly between 1 and 1000 (inclusive)
random_numbers = np.random.randint(1, 1001, 10000000)
#Counts how many times each integer appears and sorts by number (1 to 1000).
frequency = pd.Series(random_numbers).value_counts().sort_index()
# Plot the frequency
plt.figure(figsize=(15, 5))
plt.plot(frequency.index, frequency.values, marker='o', linestyle='-')
plt.xlabel('Number')
plt.ylabel('Frequency')
plt.title('Frequency of Random Integers from 1 to 9')
# plt.xticks(range(1, 10000000))
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

The plot looks like a relatively flat line, with small fluctuations around the mean frequency. Since the numbers are sampled uniformly, each number between 1 and 1000 should appear roughly the same number of times.

# Random Sample Distribution for Normal Distribution

```python
# Set random seed for reproducibility
np.random.seed(42)

# Generate data: normal distribution centered at 500 with std deviation of 100
original_data = np.random.normal(loc=500, scale=100, size=1000000)
original_data = original_data.astype(int)
# Clip to keep values between 1 and 1000
original_data = original_data[(original_data >= 100) & (original_data <= 800)]

value_range = np.arange(100, 801)  # Values from 1 to 1000
value_counts = np.bincount(original_data, minlength=801)[100:]  # Skip 0 index
prob_distribution = value_counts / value_counts.sum()

# Sample 1000 points randomly from the original data
random_sample = np.random.choice(original_data, size=100000, replace=True)

# Plotting the histograms
plt.figure(figsize=(12, 6))

# Histogram for original data
plt.subplot(1, 3, 1)
plt.hist(original_data, bins=50, color='skyblue', edgecolor='black')
plt.title('Original Data Distribution (Normal)')
plt.xlabel('Value')
plt.ylabel('Frequency')
#Original PMF
plt.subplot(1, 3, 2)
plt.plot(value_range, prob_distribution, color='blue')
plt.title('Original Probability Distribution (PMF)')
plt.xlabel('Value')
# Histogram for sampled data
plt.subplot(1, 3, 3)
plt.hist(random_sample, bins=50, color='salmon', edgecolor='black')
plt.title('Random Sample Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Step 1: Generate original data with normal distribution
np.random.seed(42)

data1 = np.random.normal(loc=100, scale=10, size=1000000)
data1 = np.clip(data1, 60, 150).astype(int)


data2 = np.random.normal(loc=150, scale=15, size=1000000)
data2 = np.clip(data2, 80, 300).astype(int)

# # Step 2: Create the probability distribution (PMF)
value_range1 = np.arange(60, 151)  # Values from 1 to 1000
value_counts1 = np.bincount(data1, minlength=151)[60:]  # Skip 0 index
prob_distribution1 = value_counts1 / value_counts1.sum()

# # Step 2: Create the probability distribution (PMF)
value_range2 = np.arange(80, 301)  # Values from 1 to 1000
value_counts2 = np.bincount(data2, minlength=301)[80:]  # Skip 0 index
prob_distribution2 = value_counts2 / value_counts2.sum()

# Step 3: Sample from this distribution
# sampled_data = np.random.choice(value_range, size=1000, p=prob_distribution)

# Step 4: Plotting
plt.figure(figsize=(12, 6))

# Original PMF
plt.subplot(1, 2, 1)
plt.plot(value_range1, prob_distribution1, color='blue')
plt.plot(value_range2, prob_distribution2, color='red')
plt.title('Original Probability Distribution (PMF)')
plt.xlabel('Value')
plt.ylabel('Probability')

plt.tight_layout()
plt.show()
```
